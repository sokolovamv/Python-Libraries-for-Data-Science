Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки
составлял 30% от всех данных, при этом аргумент random state должен быть равен 42.

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.

Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston

boston = load_boston()
# Данные о недвижимости
data = boston["data"]
# Названия признаков
feature_names = boston["feature_names"]
# массив с целевыми значениями
target = boston["target"]
```

```python
X = pd.DataFrame(data, columns=feature_names)
X
```

```python
# информация о таблице
X.info()
# В этой таблице нет пропущенных значений.
```

```python
y = pd.DataFrame(target, columns=["price"])
y
```

```python
y.info()
# В этой таблице нет пропущенных значений.
```

```python
# Разбиение данных на тренировочную и тестовую выборку
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# модель линейной регрессии
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
# обучение модели на тренировочных данных
lr.fit(X_train, y_train)
```

```python
# предсказание модели на тестовых данных
y_pred = lr.predict(X_test)

y_pred.shape
```

### Задание 2
Создайте модель под названием model с помощью класса RandomForestRegressor из модуля sklearn.ensemble. Сделайте агрумент n_estimators равным 1000, max_depth должен быть равен 12 и random_state сделайте равным 42.

Обучите модель на тренировочных данных аналогично тому, как вы обучали модель LinearRegression,

Сделайте предсказание на тестовых данных и посчитайте R2. 

Сравните с результатом из предыдущего задания.

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
# обучение модели на тренировочных данных
model.fit(X_train, y_train.values[:, 0])
```

```python
# предсказание модели на тестовых данных
y_pred_2 = model.predict(X_test)

y_pred_2.shape
```

```python
# сопоставление реальных значения с предсказанными
# массив y_pred является двумерным, переведём его в одномерный, используя метод .flatten
# массив y_pred_2 одномерный
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
    "y_pred_2": y_pred_2,
})

check_test.head(10)
```

```python
#Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.
check_test["error"] = check_test["y_pred"] - check_test["y_test"] 
check_test["error_2"] = check_test["y_pred_2"] - check_test["y_test"] 
check_test.head()
```

```python
from sklearn.metrics import r2_score 
r2_score_1=r2_score(check_test["y_pred"], check_test["y_test"]) 
print("r2_score_1:" , r2_score_1)

r2_score_2=r2_score(check_test["y_pred_2"], check_test["y_test"]) 
print("r2_score_1:" , r2_score_2)
```

**Модель RandomForestRegressor работает лучше, чем модель LinearRegression**

```python
# Обучите модель на тренировочном наборе данных
model_GSCV.fit(X_train, y_train)
```

```python
# Просмотрите параметры лучшей модели с помощью атрибута best_params_.
model_GSCV.best_params_
```

```python
# Предскажите вероятности классов с помощью полученной модели и метода predict_proba.
y_pred_proba = model_GSCV.predict_proba(X_test)
y_pred_proba
```
